## Aufgabe 5 – if
- Unter 16: "Zu jung"
- 16–17: "Fast"
- Ab 18: "Erlaubt"


```python
age = float(input("Alter: "))

ticket = input("Ticket vorhanden? (ja/nein): ")
has_ticket = ticket == "ja"

if age < 16:
    print("Kein Einlass")
elif has_ticket:
    print("Einlass erlaubt")
else:
    print("Ticket fehlt")
```

## Aufgabe 6 – if / elif / bool

Ändere das Programm:

- Passe die Altersgrenzen an
- Ändere die Texte der Ausgaben
- Probiere andere Eingaben für Ticket (ja / nein)


```python
price = float(input("Preis: "))
member = input("Mitglied? (ja/nein): ")

is_member = member == "ja"

if price < 20:
    print("Günstig")
elif is_member:
    print("Rabatt möglich")
else:
    print("Kein Rabatt")
```

## Aufgabe 7 – Eigene Regeln

Baue das Programm um:

- Ändere die Bedingungen
- Füge mindestens eine weitere Entscheidung hinzu
- Passe das Thema an (Game, Kino, Shop, Fitness, ...)


```python
name = input("Name: ")
points = float(input("Punkte (0-100): "))

if points < 30:
    level = "Noob"
elif points < 60:
    level = "Bronze"
elif points < 80:
    level = "Silber"
else:
    level = "Gold"

print(name)
print("Level:", level)
```
